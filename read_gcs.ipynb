```python
import json
import pandas as pd
from unidecode import unidecode
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession, functions as f
```

```python
sc = SparkContext.getOrCreate(conf = SparkConf())
spark = SparkSession.builder.appName("gabriel.paiva").getOrCreate()
spark.conf.set('spark.sql.repl.eagerEval.enabled', True)
spark.conf.set("spark.sql.execution.arrow.enabled","true")
```

```python
path = 'gs://cnt-landing-zone/DATA_FILE/cnt_ora_0000_kill_vendas'
```

```python
file = spark \
    .read \
    .parquet(path)
```

```python
df = file.where(
    file['dat_chave'].between(f.date_sub(f.current_date(), 16), f.date_sub(f.current_date(), 1))
).withColumn(
    'index', f.monotonically_increasing_id() + 1
)
```
