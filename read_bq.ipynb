```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('cit-teste-gcp').getOrCreate()

spark.conf.set("viewsEnabled","true")
spark.conf.set("materializationDataset","Raw")
```

```python
sql = """
  select * from `dev-data-platform-291914.Raw.cnt_xls_0000_man_meta_diaria_operacoes`
  """
df_meta_diaria_operacoes = spark.read.format("bigquery").load(sql)
df_meta_diaria_operacoes.createOrReplaceTempView("tmp_df_meta_diaria_operacoes")
```

```python
spark.sql(""" select * from tmp_df_meta_diaria_operacoes""").show()
```
